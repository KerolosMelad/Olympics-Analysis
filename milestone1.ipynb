# 2 - Data Preprocessing and Cleaning

# 2-a Imputation

```python
imputed_df= mergerd_df.copy()
```

It seems that most of the notes are missing in addition it is not informative, so I decided to dispose of this useless feature.

```python
imputed_df.drop('notes' ,axis = 1, inplace= True)
```

I had to encode some columns to visualize the relations between the columns I wish to impute and other categeorical columns 

```python
cat_col= imputed_df[['Sex','Season' , 'Medal' ]]
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore',sparse=False)
coded_col= pd.DataFrame(encoder.fit_transform(cat_col))
coded_col.index = imputed_df.index
coded_col.columns = encoder.get_feature_names_out(cat_col.columns)
imputed_df = pd.concat([imputed_df.drop(cat_col.columns, axis=1), coded_col], axis=1)
imputed_df = imputed_df.rename(columns={'Medal_nan': 'No_Medal'})
```

Take into consedration that medal missing values in Medals column are MNAR .. So I assumed that all nan valeus the ones got no medals

```python
gold_counts = imputed_df['Medal_Gold'].sum()
silver_counts = imputed_df['Medal_Silver'].sum()
bronze_counts = imputed_df['Medal_Bronze'].sum()
print(gold_counts,silver_counts,bronze_counts)
```

We can notice that the number of medals are not equal, However the difference is too small that we can ignore 

```python
plt.figure(figsize=(12,12))
sns.heatmap(data=imputed_df.corr(),annot=True)
```

Age is not correlated to one of the features except for the weight, height and sex. However, it 's not highly correlated and these columns has to be processed

so we can't depend on them while imputing the age 

I found it reasonable to map the age with median age of the corresponding sport 

```python
plt.figure(figsize=(8,4))
sns.kdeplot(x=imputed_df["Age"])
plt.title("Age Distribution")
```

I used the median because the distribution is skewed in general 

```python
imputed_df['Age'] = imputed_df['Age'].fillna(imputed_df['Sport'].map(imputed_df.groupby('Sport')['Age'].median()))
```

I found it reasonable  to map the region with mode region of the corresponding City 

```python
imputed_df['region'] = imputed_df['region'].fillna(imputed_df['NOC'].map(imputed_df.groupby('NOC')['region'].agg(pd.Series.mode)))
```

Weight and Height are highly correlated to each other and to sex and age 

I decided to impute it using linear regression


```python
# Weight imputation 
from sklearn.linear_model import LinearRegression
data = imputed_df.dropna(subset = ['Weight', 'Height'] ,axis =  0 )
X = data[['Age','Sex_M','Height']]
y = data['Weight']
W_reg = LinearRegression().fit(X.values, y.values)
W_reg.score(X.values, y.values)
```

However I couldn't impute weight with nan values of height so whenever there's a nan value in height feature .. I replaced it with height mean .. 

```python
plt.figure(figsize=(8,4))
sns.kdeplot(x=imputed_df["Height"])
plt.title("Height Distribution")
```

```python
import math
height_mean = imputed_df['Height'].median()
for i, row in imputed_df.iterrows():
    if math.isnan(row['Weight']):
        x=[row['Age'] , row['Sex_M'] ]
        if  math.isnan(row['Height']):
            x = x+[height_mean]
        else:
            x = x+[row['Height']]
            
        predicted = W_reg.predict([x])
        imputed_df.at[i,'Weight'] = predicted
```

```python
data = imputed_df.dropna(subset = ['Height'] ,axis =  0 )
X = data[['Age','Sex_M','Weight']]
y = data['Height']
W_reg = LinearRegression().fit(X.values, y.values)
W_reg.score(X.values, y.values)
```

```python
import math
for i, row in imputed_df.iterrows():
    if math.isnan(row['Height']):
        x=[row['Age'] , row['Sex_M'] ,row['Weight'] ]
        predicted = W_reg.predict([x])
        imputed_df.at[i,'Height'] = predicted
```

Here we defined a function that determines the outlier values for a specific column based on its boxplot and the whisker values

```python
def IQR_technique(column,df):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    cut_off = IQR * 1.5
    lower = Q1 - cut_off
    upper =  Q3 + cut_off
    return lower ,upper
```
